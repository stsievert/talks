```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import binom
```

```python
x = np.arange(100)
y = binom(x, 3)

plt.figure()
plt.plot(x, y)
plt.xlabel('Number of faces')
plt.ylabel('Number of possible questions')
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
```

```python
x = np.arange(100)
y1 = binom(x, 3)
y2 = binom(x, 2)

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('Number of faces')
plt.ylabel('Number of possible questions')
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
```
